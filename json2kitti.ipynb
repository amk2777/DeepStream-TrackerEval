```python
import json
```

```python
with open('formated_KITTI_cummulated-v5.json') as f:
    data = json.load(f)
```

```python
data
```

```python
# {frame_id : [], tracker_id: [label], truncated :[], occluded:[], alpha:[], bbox:[x1,y1,x2,y2], dimensions:[], location:[], rotation_y:[], score:[]  }
```

```python
kitti_lines = []
```

```python
for id,label in data.items():
#     for prop in label.items():
    frame_id = label['filename'].split('.')[0]
    frame_id = int(frame_id.split('_')[-1])
    for region in label['regions']:
        tracker_id = region['region_attributes']['tracker id']
        class_id = region['region_attributes']['class']
        left=region['shape_attributes']['x']
        top=region['shape_attributes']['y']
        right=region['shape_attributes']['width']
        bottom=region['shape_attributes']['height']
        
#        kitti_lines.append(f'{frame_id} {tracker_id} {class_id} {0} {0} {0} {left} {top} {right} {bottom} {0} {0} {0} {0} {0} {0} {0} {-0.100000}\n')
        kitti_lines.append([frame_id, tracker_id, class_id, 0, 0, 0, left, top, right, bottom, 0, 0, 0, 0, 0, 0, 0, -0.100000])
#     print(f'frame: {frame_id} and label: {label}')
```

```python
import pandas as pd
import numpy as np
```

```python
kitti = np.array(kitti_lines)
```

```python
df = pd.DataFrame(kitti)
```

```python
df = df.sort_values(0)
```

```python
df = df.astype({0: int})
```

```python
df
```

```python
df=df.drop_duplicates(subset=[0, 1], keep='last')
df
```

```python
df.to_csv('tracker-labels.txt', sep=' ', header=False, index=False)
```
